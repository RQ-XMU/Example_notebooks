```python
# list of batches of size = 20
train_dataset = []

batch_size = 20

# split the trasnform data into batches of 20
# transform_data size (10000, 1) = vector of 10.000 names
# train_dataset will have shape (batch num, char num in name, batch size, one-hot size) = 
# 500 x 12 x 20 x 27
# 12 = number of step in LSTM


for i in range(len(transform_data)-batch_size+1):
#     print(i)
    start = i*batch_size
    end = start+batch_size
    
    #batch data
    batch_data = transform_data[start:end]
    
    if(len(batch_data)!=batch_size):
        break  # the last batch was reached
        
    #convert each char of each name of batch data into one hot encoding
    char_list = []
    for k in range(len(batch_data[0][0])): #(k = 0...char_num -1)
        batch_dataset = np.zeros([batch_size,len(vocab)])
        for j in range(batch_size):
            name = batch_data[j][0]
            char_index = char_id[name[k]]  # get k-th char in name, transfer to index in vocab
            batch_dataset[j,char_index] = 1.0  # then enconde = one-hot
     
        #store the ith char's one hot representation of each name in batch_data
        char_list.append(batch_dataset)
    
    #store each char's of every name in batch dataset into train_dataset
    train_dataset.append(char_list)
```

```python
batch_size = 20

batch_numbers = len(transform_data)//batch_size
if batch_size * batch_numbers < len(transform_data):
    batch_numbers +=1
    
max_name_len = max_len
one_hot_dimenssion = len(vocab)
train_data = np.zeros((batch_numbers, max_name_len, batch_size, one_hot_dimenssion))

for batch_id in range(batch_numbers):
    start_batch_id = batch_id * batch_size
    end_batch_id = start_batch_id + batch_size
    
    batch_train = transform_data[start_batch_id: end_batch_id]
    
    for name_id in range(batch_size):
        name = batch_train[name_id][0]
        
        for char_idd in range(max_name_len):
            char = list(name)[char_idd]
            
            first = batch_id
            second = char_idd
            third = name_id
            fourth = char_id[char]
            
            train_data[first, second, third, fourth] = 1
            
```

## Reference: *Adam optimizer*<br>
Adam = *adaptive moment estimation*<br>
===================================<br>
$\textbf{Require}: \alpha - \text{step size}$: learning rate<br>
$\textbf{Require}: \beta_1, \beta_2 \in \left[0,1\right)$ - Exponential decay rates for moment estimates<br>
$\textbf{Require}: f(\theta):$ - stochastic objective function with parameter $\theta$<br>
$\textbf{Require}: \theta_0:$ - initial parameter vector<br>
The goal is update parameters such that minimize objective function.<br>
$01.\ \ \ m_0 \leftarrow 0:$ initiate first moment vector<br>
$02.\ \ \ v_0 \leftarrow 0:$ initiate second moment vector<br>
$03.\ \ \ t \leftarrow 0:$ initiate time step<br>
$04.\ \ \textbf{While}\ \theta_t\ \text{not converged}\ \textbf{do:}$<br>
$05.\ \quad t\leftarrow t+1$<br>
$06.\ \quad g_t \leftarrow\triangledown_{\theta} f_t(\theta_{t-1})$: get gradients w.r.t. stochastic objective at time step $t$ <br>
$07.\ \quad m_t \leftarrow \beta_1 \cdot m_{t-1} + (1-\beta_1)\cdot g_t$: update biased first moment estimate<br>
$08.\ \quad v_t \leftarrow \beta_2\cdot v_{t-1} + (1-\beta_2)\cdot g_t^2$: Update biased second raw moment estimate<br>
$09.\ \displaystyle \quad \hat{m}_t \leftarrow \frac{m_t}{1-\beta_1^t}$: Compute bias-corrected first moment estimate<br>
$10.\ \displaystyle \quad \hat{v}_t \leftarrow \frac{v_t}{1-\beta_2^t}$: Compute bias-corrected second raw moment estimate<br>
$11.\ \displaystyle \quad \theta_t \leftarrow \theta_{t-1}-\alpha\cdot\frac{\hat{m}_t}{\sqrt{\hat{v}_t}+\varepsilon}$: Update parameters<br>
$12.\ \ \ \textbf{end while}$<br>
$13.\ \ \ \textbf{Retrun}\ \theta_t$: return parameters<br>
===================================<br>

## Reference: *Activation functions*
**Sigmoid function:**
$$\sigma(x)=\frac{1}{1+e^{-x}}$$
$$\frac{\partial}{\partial x}\sigma(x) = \sigma(x)[1-\sigma(x)]$$

**Tanh function:**
$$\tanh(x) = \frac{e^x-e^{-x}}{e^x+e^{-x}}$$
$$\frac{\partial}{\partial x}\tanh(x)=1-\tanh^2(x)$$

**ReLU function:** (usually used in CNNs)
$$ReLU(x)=\max(0,x)$$
$$\frac{\partial}{\partial x}ReLU(x)=\frac{ReLU(x)}{x}$$

**Softmax function:** - a generative form of **Sigmoid**<br>
variable $X=(x_1, x_2,..., x_n)$, function output $Y=(y_1, y_2,..., y_n)$, <br>
where:
$$y_i=\frac{e^{x_i}}{\sum_{j=1}^{n}{e^{x_j}}}$$
In vector form:
$$Y=\frac{e^X}{\sum{e^X}}$$
Derivative has the same form of **sigmoid**:
$$\frac{\partial y_i}{\partial x_i}=y_i(1-y_i)$$

```python
#Activation Functions
#sigmoid
def sigmoid(X):
    return 1/(1+np.exp(-X))

#tanh activation
def tanh_activation(X):
    return np.tanh(X)

#softmax activation
def softmax(X):
    exp_X = np.exp(X)
    exp_X_sum = np.sum(exp_X,axis=1).reshape(-1,1)
    exp_X = exp_X/exp_X_sum
    return exp_X

# ReLU activation
def relu(X):
    return np.clip(X, a_min=0, a_max=None)

#derivative of tanh
def tanh_derivative(X):
    return 1-(X**2)
```

### The Core Idea Behind LSTMs
source: http://colah.github.io/posts/2015-08-Understanding-LSTMs/
<br>
The key to LSTMs is the cell state, the horizontal line running through the top of the diagram.<br>
The cell state is kind of like a conveyor belt. It runs straight down the entire chain, with only some minor linear interactions. It’s very easy for information to just flow along it unchanged.

```python
Image.open('LSTM3-C-line.png')
```

The LSTM does have the ability to remove or add information to the cell state, carefully regulated by structures called gates.

Gates are a way to optionally let information through. They are composed out of a sigmoid neural net layer and a pointwise multiplication operation.

```python
Image.open('LSTM3-gate.png')
```

The **sigmoid layer** outputs numbers between zero and one, describing how much of each component should be let through. A value of *zero* means “let nothing through,” while a value of *one* means “let everything through!”

An LSTM has three of these gates, to protect and control the cell state.
### Step-by-Step LSTM Walk Through
The first step in our LSTM is to decide *what information* we’re going *to throw* away from the cell state. This decision is made by a sigmoid layer called the “*forget gate layer*.” It looks at $h_{t−1}$ and $x_t$, and outputs a number between $0$ and $1$ for each number in the cell state $C_{t−1}$. A $1$ represents “completely keep this” while a $0$ represents “completely get rid of this.”

```python
Image.open('LSTM3-focus-f.png')
```

$[h_{t-1}, x_t]$ means to concatenate previous cell output $h_{t-1}$ with current cell input $x_t$

The next step is to decide *what new information* we’re going *to store* in the cell state. This has two parts. First, a **sigmoid layer** called the “input gate layer” decides which values we’ll *update*. Next, a **tanh layer** creates a vector of new candidate values, $\tilde{C}_t$, that could be added to the state. In the next step, we’ll *combine these two to create an update* to the state.

```python
Image.open("LSTM3-focus-i.png")
```

It’s now time to update the old cell state, $C_{t−1}$, into the new cell state $C_t$. The previous steps already decided what to do, we just need to actually do it.

We *multiply* the old state *by* $f_t$, forgetting the things we decided to forget earlier. Then we add $i_t∗\tilde{C}_t$. This is the new candidate values, scaled by how much we decided to update each state value.

```python
Image.open("LSTM3-focus-C.png")
```

Finally, we need to decide *what* we’re going to *output*. This output will be based on our *cell state*, but will be a filtered version. First, we run a **sigmoid layer** which decides *what parts* of the cell state we’re going *to output*. Then, we put the cell state through **tanh** (to push the values to be between −1 and 1) and multiply it by the output of the sigmoid gate, so that we only output the parts we decided to.

```python
Image.open("LSTM3-focus-o.png")
```

## Initialize Parameters
Embeddings Size = **100**<br>
Hidden Units (size) = **256**<br>
Total INPUT Weights (size) = 100 + 256 = **356** <br>
### LSTM CELL Weights
- Forget Gate Weights (size) = {356,256} ($W_f$)
- Input Gate Weights (size) = {356,256} ($W_i$)
- Gate Gate Weights (size) = {356,256} ($W_c/W_g$)
- Output Gate Weights (size) = {356,256} ($W_o$)

### Output CELL Weights
- Output Weights (size) = {256,27}

Store these weights in parameters dictionary!

```python
#initialize parameters
def initialize_parameters():
    #initialize the parameters with 0 mean and 0.01 standard deviation
    mean = 0
    std_cell = 2/np.sqrt(input_units+hidden_units+hidden_units)
    
    #lstm cell weights loc=0.0, scale=1.0, size=None
    forget_gate_weights = np.random.normal(mean,std_cell,(input_units+hidden_units,hidden_units))
    input_gate_weights  = np.random.normal(mean,std_cell,(input_units+hidden_units,hidden_units))
    output_gate_weights = np.random.normal(mean,std_cell,(input_units+hidden_units,hidden_units))
    gate_gate_weights   = np.random.normal(mean,std_cell,(input_units+hidden_units,hidden_units))
    
    #hidden to output weights (output cell)
    std_out = 2/np.sqrt(hidden_units+output_units)
    hidden_output_weights = np.random.normal(mean,std_out,(hidden_units,output_units))
    
    parameters = dict()
    parameters['fgw'] = forget_gate_weights
    parameters['igw'] = input_gate_weights
    parameters['ogw'] = output_gate_weights
    parameters['ggw'] = gate_gate_weights
    parameters['how'] = hidden_output_weights
    
    return parameters
```

### LTSM Cell

```python
Image.open('LSTM-cell.JPG')
```

#### Equations
- fa = sigmoid(fgw . [xt, at-1])
- ia = sigmoid(igw . [xt, at-1])
- ga = tanh(ggw . [xt, at-1])
- oa = sigmoid(ogw . [xt, at-1])
- ct = ia * ga + fa * ct-1
- at = oa * tanh(ct)

```python
def lstm_cell(batch_dataset, prev_activation_matrix, prev_cell_matrix, parameters):
    """
    Single LSTM cell activator
    :param: batch_dataset, input = xt, char_num x batch_size x embedding_size
    :param: prev_activation_matrix = at-1, batch_size x hidden_units
    :param: prev_cell_matrix = ct-1, batch_size x hidden_units
    :param: parameters: dictionary of weight matrices {fgw, igw, ogw, ggw, how}
    :return: LSTM activation (fa, ia, ga, oa), cell matrix (ct) & activation matrix (at)
    """
    
    # get weight matrices
    fgw = parameters['fgw']  # (input_units + hidden_units) x hidden_units = 356x256
    igw = parameters['igw']  # 356x256
    ogw = parameters['ogw']  # 356x256
    ggw = parameters['ggw']  # 356x256
    
    # concat batch data (xt) and previous activation matrix (at-1)
    # batch_size x embedding_size + batch_size x hidden_units = batch_size x (embedding_size + hidden_units)
    # 20 x 100 + 20 x 256 = 20 x 356
    concat_dataset = np.concatenate((batch_dataset, prev_activation_matrix), axis=1)
    
    # forget gate activation
    fa = np.matmul(concat_dataset, fgw)
    fa = sigmoid(fa)  # 20x256
    
    # input gate activation
    ia = np.matmul(concat_dataset, igw)
    ia = sigmoid(ia)  # 20x256
    
    # ouput gate activation
    oa = np.matmul(concat_dataset, ogw)
    oa = sigmoid(oa)  # 20x256
    
    # gate gate activation
    ga = np.matmul(concat_dataset, ggw)
    ga = tanh_activation(ga)  # 20x256
    
    # new cell memory matrix (ct)
    cell_memory_matrix = np.multiply(fa, prev_cell_matrix) + np.multiply(ia, ga)  # 20x256
    
    # new activation matrix (at)
    activation_matrix = np.multiply(oa, tanh_activation(cell_memory_matrix))  # 20x256
    
    # store activations for next back prop
    lstm_activations = {}
    
    lstm_activations['fa'] = fa
    lstm_activations['ia'] = ia
    lstm_activations['oa'] = oa
    lstm_activations['ga'] = ga
    
    return lstm_activations, cell_memory_matrix, activation_matrix
```

### Output cell

Equations

- ot = softmax(at * how)

```python
def output_cell(activation_matrix, parameters):
    """
    Get output (ht)
    :param: activation_matrix: (= at) # 20x256
    :param: parameters: dict of weight matrices (fgw, igw, ogw, ggw, how)
    """
    
    # get hidden to output weight matrix
    how = parameters['how']  # 256x27
    
    output = np.matmul(activation_matrix, how)
    output = softmax(output)
    
    return output  # 20x27, batch_size x one-hot_size
```

### Get corresponding embeddings for the batch dataset

```python
def get_embeddings(batch_dataset, embeddings):
    """
    Transform batch data to embeddings data
    :param: batch_dataset: batch data set, batch_size x one-hot_size (20x27)
    :param: embeddings - embedding matrix, one-hot_size x embedding_size (27x100)
    :return: embedded batch dataset, batch_size x embedding_size (20x100)
    """
    
    embedded_dataset = np.matmul(batch_dataset,embeddings)
    
    return embedded_dataset
```

### Forward Propagation

```python
Image.open("LSTMForward.JPG")
```

Function returns the intermediate ativations in the respective caches:

- LSTM Cache :- All LSTM cell activation in every cell [fa, ia, ga, oa],...
- Activation Cache : All activation (a0, a1, a2...)
- Cell Cache : All cell activations (c0, c1, c2...)
- Embedding cache : Embeddings of each batch (e0, e1, e2...)
- Output Cache : All output (o1, o2, o3...)

```python
def forward_propagation(batches, parameters, embeddings):
    """
    Forward propagation
    :param batches: train data in batches, size = char_num(12)xbatch_size(20)xone-hot_size(27) 
    :param parameters: weight matrices: igw, ogw, fgw, ggw, how
    :param embeddings: embeding matrix, shape = one-hot_size(27) x embed_size(100)
    :return:
        - embedding_cache, 'emb0'...'emb11', each shaped batch_size(20) x embed_size(100)
              embedding_cache['emb0'] = temporary embedding vectors of 20 first characters in batch
              embedding_cache['emb0'] = temporary embedding vectors of 20 second characters in batch
        
        - lstm_cache: 'lstm1' ... 'lstm11', 1st, 2nd, ..., 11th of activation tuple('fa', 'ia', 'oa', 'ga')
              lstm_cache['lstm1'] = activations in first hidden cell, each shaped batch_size(20) x hidden_units (256)
        
        - activation_cache: 'a0', 'a1', ..., 'a11': activation caches, input of output layer, 
            activation_cache['a0'] = activation of first cell, shaped batch_size(20) x hidden_units (256)
        
        - cell_cache: 'c0', 'c1', ..., 'c11': cell caches, each shaped batch_size x hidden_units
            cell_cache['c0'] - cache of firts cell (ct0)
            
        - output_cache: 'o1', 'o2', 'o11', output of each cell, each shaped batch_size(20)xone-hot_size(27)
          softmax of activation_caches, 'o1'[i,j] = probability that 
          1st character in i-th data point is char with index j in vocabulary'o1'
    """
    
    #get batch size
    batch_size = batches[0].shape[0] # batches (500, 12, 20, 27) => batch size = 20
    
    # for storing the activations of all the unrollings.
    lstm_cache = dict()                 # lstm cache
    activation_cache = dict()           # activation cache 
    cell_cache = dict()                 # cell cache
    output_cache = dict()               # output cache
    embedding_cache = dict()            # embedding cache
    
    #initial activation_matrix(a0) and cell_matrix(c0)
    a0 = np.zeros([batch_size,hidden_units],dtype=np.float32)
    c0 = np.zeros([batch_size,hidden_units],dtype=np.float32)
    
    #store the initial activations in cache
    activation_cache['a0'] = a0
    cell_cache['c0'] = c0
    
    #unroll the names
    for i in range(len(batches)-1):
        
        # get first character batch
        batch_dataset = batches[i] # (20, 27)
        
        # get embeddings => (20, 100)
        batch_dataset = get_embeddings(batch_dataset=batch_dataset, embeddings=embeddings)
        embedding_cache['emb'+str(i)] = batch_dataset
        
        # lstm cell activations
        lstm_activations,ct,at = lstm_cell(
            batch_dataset=batch_dataset, 
            prev_activation_matrix=a0, 
            prev_cell_matrix=c0, 
            parameters=parameters
        )
        
        # output cell, 20x27
        ot = output_cell(activation_matrix=at, parameters=parameters)
        
        # store the time 't' activations in caches
        lstm_cache['lstm' + str(i+1)]  = lstm_activations
        activation_cache['a'+str(i+1)] = at
        cell_cache['c' + str(i+1)] = ct
        output_cache['o'+str(i+1)] = ot
        
        #update a0 and c0 to new 'at' and 'ct' for next lstm cell
        a0 = at
        c0 = ct
        
    return embedding_cache, lstm_cache, activation_cache, cell_cache, output_cache
```

```python
embeddings = np.random.normal(0,0.01,(len(vocab),input_units))
batches = train_dataset[0]
parameters = initialize_parameters()
fwp = forward_propagation(batches,parameters,embeddings)
```

### Calculate the Loss, Perplexity, and Accuracy
#### Loss:
- Loss at time t: $Loss(t) = -\frac{1}{m}\sum_{i=1}^{m}[\text{label}^{(i)}\log(\text{prediction}^{(i)})+(1-\text{label}^{(i)})\log(1-\text{prediction}^{(i)})]$
- Overall loss: $Loss = \sum_{t}Loss(t)$ 

#### Perplexity:
- Probability Product = $\prod_{i=1}^{char\_num}Probability(predicted\ char)$
- Perplexity = $(Probability\ Product)^{1/char\_num}$

#### Accuracy:
- Accuracy(t) = (Y==predictions, axis=1) for all time steps
- Accuracy = ((∑Acc(t))/batch_size)/n for all time steps, n is number of chars in name

```python
def cal_loss_accuracy(batch_labels, output_cache):
    """
    Calculate loss, perplexcity and calculate
    :param batch_labels: labels for batches, 12 x 20 x 27
    :param output_cache: cache of ht [o0, o1, o2, ...] 11 x 20 x 27
    """
    
    loss = 0  # to sum loss for each time step
    acc = 0   # to sum accuracy for each time step
    prob = 1  # probability product of each time step predicted char
    
    # batch size
    batch_size = batch_labels[0].shape[0] # 20
    
    # loop through each time step
    for i in range(1, len(output_cache)+1):
        
        # get true labels and predictions
        labels = batch_labels[i] # 20x27
        pred = output_cache['o'+str(i)] # 20x27
        
        # update loss, product probability, accuracy each time step t, # 20x1
        prob = np.multiply(prob, np.sum(np.multiply(labels, pred), axis=1).reshape(-1,1)) 
        loss += np.sum((np.multiply(labels, np.log(pred)) + np.multiply(1-labels, np.log(1-pred))), axis=1).reshape(-1,1)
        acc = np.array(np.argmax(labels,1)==np.argmax(pred,1), dtype=np.float32).reshape(-1,1)
        
    # correct loss, perplexity, accuracy =. down to number
    loss = np.sum(loss) * (-1.0/batch_size)
    perplexity = np.sum((1/prob)**(1/len(output_cache)))/batch_size
    acc = np.sum(acc)/batch_size
    acc = acc/len(output_cache)
    
    return perplexity, loss, acc
    
```


#### Calculate Output Cell Errors for each time step

- Output Error Cache :- to store output error for each time step
- Activation Error Cache : to store activation error for each time step

Errors are used for backpropagation

```python
def calculate_output_cell_error(batch_labels, output_cache, parameters):
    """
    Calculate output cell errors (errors of h1, h2, h3, ...)
    :param batch_labels: label batches
    :param output_cache: o1, o2, o3 ...
    :param parameters: weight matrices
    """
    
    #for storing the output errors for each time step
    how = parameters['how']
    output_error_cache = dict()
    activation_error_cache = dict()
    
    # loop through each time step
    for i in range(1, len(output_cache)+1):
        
        #get true and predicted labels
        labels = batch_labels[i]
        pred = output_cache['o'+str(i)]
        
        #calculate the output_error for time step 't'
        error_output = pred - labels
        
        #calculate the activation error for time step 't'
        error_activation = np.matmul(error_output,how.T)
        
        #store the output and activation error in dict
        output_error_cache['eo'+str(i)] = error_output
        activation_error_cache['ea'+str(i)] = error_activation
        
    return output_error_cache, activation_error_cache
```
